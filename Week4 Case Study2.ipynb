# 1-Read the Given Dataset to Python Environment

```python
import pandas as pd
cars = pd.read_csv('cars_data.csv')
cars
```

# 2-Null Values in the Dataset

```python
cars.info()
```

# 3-Countplot of Male vs  female buyers participated in the sales

#### Here we displayed the count plot of male vs female buyers participated in the sales and applied grid to the plot for better understanding.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))

sns.countplot(cars['Buyer Gender'])
plt.title('Count Of Male vs Female Buyers In The Sale',fontsize=16)
plt.grid()
```

# 4-Finding Top 5 Cars Based on Their Sales Price

#### Here We Sort The Given Data Based On The Sale Price In Descending Order And Displayed The Details Of Top 5 Cars

```python
cars_top = cars.sort_values('Sale Price',ascending=False).head(5)
cars_top
```

### Here We Assign The Car Names Into A New Variable As Series And Changed Its Index As Their Order,Then Displayed It.

```python
print('The Top 5 Car Making Companies Based on Their Sales Price Are')
top = pd.Series(cars_top['Make'])
top.index=[1,2,3,4,5]
top
```

# 5-Finding Least 5 Cars Based on Their Resell Price

#### Here We Sort The Given Data Based On The Resell Price In Ascending Order And Displayed The Details Of Least 5 Cars

```python
cars_least = cars.sort_values('Resell Price').head(5)
cars_least
```

### Here We Assign The Car Names Into A New Variable As Series And Changed Its Index As Their Order,Then Displayed It.

```python
print('The Least 5 Car Making Companies Based on Their Resell Price Are')
least = pd.Series(cars_least['Make'])
least.index=[1,2,3,4,5]
least
```
